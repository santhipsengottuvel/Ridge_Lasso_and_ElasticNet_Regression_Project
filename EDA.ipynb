## Data Cleaning

```python
## Checking for null values
df.isnull().sum()
```

```python
## Creating a column 'Region' and combining the two region data
df.loc[:122,"Region"]=str(0)
df.loc[122:,"Region"]=str(1)
```

```python
## Checking the 122 th index whether the dataset is combined correctly or not
df.loc[122]
```

```python
## Droping down the headings of the second columns
df=df.drop(122).reset_index(drop=True)
```

```python
df.head()
```

```python
## Removing the spaces in the column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
## Chenging the datatypes of the columns
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the datatypes of the columns
object=[feature for feature in df.columns if df[feature].dtypes=='O']
for i in object:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
## Monthly Fire Analysis
## Renaming encoded classification feature back to normal for better visualization
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'Not Fire','Fire')
```

```python
df_region1=df.loc[df['Region']==0]
df_region2=df.loc[df['Region']==1]
```

```python
## Countplot to visualize fire analysis based on months for Bejaia Region
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data=df_region1,hue='Classes',x='month')
plt.title("Fire Analysis of Bejaia Region by Month",weight='bold',style='italic')
plt.xlabel("Months",weight='bold')
plt.ylabel("Count of Fires",weight='bold')
```

```python
## Countplot to visualize fire analysis based on months for Sidi-Bel Abbes Region
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data=df_region2,hue='Classes',x='month')
plt.title("Fire Analysis of Sidi-Bel Abbes by Month",weight='bold',style='italic')
plt.xlabel("Months",weight='bold')
plt.ylabel("Count of Fires",weight='bold')
```

```python
## Encoding the categorical features again for model fitting
df['Classes']=np.where(df['Classes'].str.contains('Not Fire'),0,1)
```
